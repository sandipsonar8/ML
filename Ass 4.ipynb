```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({
    "Date":pd.date_range(start="2021-10-01",periods=10,freq="D"),
    "Item":1014,
    "M1":np.random.randint(1,10,size=10),
    "M2":np.random.random(10).round(2),
    "M3":np.random.random(10).round(2),
    "M4":np.random.random(10)
})
df
```

```python
df.loc[[2,9],"Item"]=np.nan
df.loc[[2,7,9],"M1"]=np.nan
df.loc[[2,3],"M2"]=np.nan
df.loc[[2],"M3"]=np.nan
df.loc[:6,"M4"]=np.nan
df
```

```python
df=df.astype({
    "Item":pd.Int64Dtype(),
    "M1":pd.Int64Dtype()
})
df
```

```python
#1. Drop rows or columns that have a missing value
df.dropna()
```

```python
df.dropna(axis=1)
```

```python
df.dropna(how='all')
```

```python
#2. Drop rows or columns based on a threshold value
df.dropna(thresh=4)
```

```python
#3
df.dropna(subset=["M2","M3"])
```

```python
#4
values={"Item":1014,"M1":0}
df.fillna(value=values)
```

```python
#Handling Missing Values
import pandas as pd
data=pd.read_csv('abc.csv')
data
```

```python
print("mean()=>")
data['price']=data['price'].fillna(data['price'].mean())
print(data)
```

```python
print("max()=>")
data['price']=data['price'].fillna(data['price'].max())
print(data)
```

```python
print("min()=>")
data['price']=data['price'].fillna(data['price'].min())
print(data)
```

```python
print("std()=>")
data['price']=data['price'].fillna(data['price'].std())
print(data)
```
