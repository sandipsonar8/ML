```python
# Generate sample data
from sklearn.datasets import make_blobs
X, Y = make_blobs(n_samples=50, centers=5, random_state=42)

# Create the Agglomerative Clustering model with 2 clusters
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(n_clusters=2)
model.fit(X)

# Get the cluster labels for each data point
cluster_label= model.labels_

import matplotlib.pyplot as plt
# Plot the data points and cluster assignments
plt.scatter(X[:, 0], X[:, 1],c=cluster_label)
plt.title("Agglomerative Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
iris=datasets.load_iris()
X=iris.data
Y=iris.target
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plot
X=PCA(n_components=2).fit_transform(X)
plot.scatter(X[:,0],X[:,1])
```

```python
#The Agglomerative clustering module present inbuilt in sklearn is used for this purp
from sklearn.cluster import AgglomerativeClustering
classifier = AgglomerativeClustering(n_clusters = 3)
clusters = classifier.fit_predict(X)
```

```python
#Visualize the results of the clustering algorithm.
plt.scatter(X[clusters == 0, 0], X[clusters == 0, 1], label = 'Type 1')
plt.scatter(X[clusters == 1, 0], X[clusters == 1, 1], label = 'Type 2')
plt.scatter(X[clusters == 2, 0], X[clusters == 2, 1], label = 'Type 3')
plt.title('Clusters')
plt.show()
```
