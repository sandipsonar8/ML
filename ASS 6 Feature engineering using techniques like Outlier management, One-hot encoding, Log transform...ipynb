```python
import pandas as pd
df=pd.read_csv('team.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data=df
data.TEAM=le.fit_transform(data.TEAM)
data
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
model=OneHotEncoder()
model_data=pd.DataFrame(model.fit_transform(data[['TEAM']]).toarray())
model_data
```

```python
data=data.join(model_data)
data
```
